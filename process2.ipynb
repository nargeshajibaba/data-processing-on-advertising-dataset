```python
d = []
for i in df["Age"]:
    if i>=18 and i<=40:
        d.append("young")
    elif i>40 and i<=60:
        d.append("middle-aged")
    elif i>60:
        d.append("old")
df["Categorize age"] = d
```

```python
df
```

```python
young = df[df["Categorize age"] == "young"]
middle = df[df["Categorize age"] == "middle-aged"]
old = df[df["Categorize age"] == "old"]
```

```python
plt.pie([len(young) , len(middle) , len(old)] , labels=["young" , "middle-aged" , "old"])
plt.show()
```
